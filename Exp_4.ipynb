```python
# Given data
data = [10, 20, 20, 30, 40]
n = len(data)

# Mean
mean = sum(data) / n

# Median
data.sort()
if n % 2 == 0:
    median = (data[n//2 - 1] + data[n//2]) / 2
else:
    median = data[n//2]

# Mode
mode = data[0]
max_count = 0
for x in data:
    count = 0
    for y in data:
        if x == y:
            count += 1
    if count > max_count:
        max_count = count
        mode = x

# Variance
variance = 0
for x in data:
    variance += (x - mean) ** 2
variance = variance / n

# Standard Deviation
std_dev = variance ** 0.5

# Output
print("Data:", data)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Get data from user input
user_input = input("Enter numbers separated by spaces or commas: ")
# Split by commas first, then by spaces, and filter out empty strings
data = [float(x.strip()) for x in user_input.replace(' ', ',').split(',') if x.strip()]

n = len(data)

# Mean
mean = sum(data) / n

# Median
data.sort()
if n % 2 == 0:
    median = (data[n//2 - 1] + data[n//2]) / 2
else:
    median = data[n//2]

# Mode
# A more robust way to calculate mode for multiple occurrences
from collections import Counter
counts = Counter(data)
max_count = max(counts.values())
mode = [key for key, value in counts.items() if value == max_count]
# If there's only one mode, extract it from the list
if len(mode) == 1:
    mode = mode[0]

# Variance
variance = 0
for x in data:
    variance += (x - mean) ** 2
variance = variance / n

# Standard Deviation
std_dev = variance ** 0.5

# Output
print("Data:", data)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Get user input for mean and standard deviation
try:
    mean_input = float(input("Enter the mean (e.g., 0): "))
    std_dev_input = float(input("Enter the standard deviation (e.g., 1): "))
except ValueError:
    print("Invalid input. Please enter numerical values.")
    # Use default values or exit if input is invalid
    mean_input = 0
    std_dev_input = 1
    print(f"Using default mean={mean_input} and std_dev={std_dev_input}")

# Generate normal distribution data based on user input
data = np.random.normal(loc=mean_input, scale=std_dev_input, size=1000)

# Plot histogram
plt.hist(data, bins=30, density=True)

# Plot normal distribution curve
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mean_input, std_dev_input))

# Labels and title
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Histogram of Normal Distribution (User Input)")

plt.show()
```

```python
import random
from collections import Counter

def coin_toss():
    """Simulates a single coin toss and returns 'Heads' or 'Tails'."""
    return random.choice(['Heads', 'Tails'])

num_tosses = int(input("Enter the number of coin tosses to simulate: "))

toss_results = []
for _ in range(num_tosses):
    toss_results.append(coin_toss())

# Count the occurrences of Heads and Tails
toss_counts = Counter(toss_results)

heads_count = toss_counts.get('Heads', 0)
tails_count = toss_counts.get('Tails', 0)

# Calculate probabilities
probability_heads = heads_count / num_tosses
probability_tails = tails_count / num_tosses

print(f"\n--- Simulation Results ---")
print(f"Total Tosses: {num_tosses}")
print(f"Heads Count: {heads_count}")
print(f"Tails Count: {tails_count}")
print(f"Probability of Heads: {probability_heads:.4f}")
print(f"Probability of Tails: {probability_tails:.4f}")
```
