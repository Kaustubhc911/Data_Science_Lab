```python
#Code to compute factorial of number
def factorial(n):
    """Calculates the factorial of a non-negative integer n."""
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")

# Another example:
number = 0
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
def display_multiplication_table(number, limit=10):
    """Displays the multiplication table for a given number up to a specified limit."""
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num = 7
display_multiplication_table(num)

print("\n") # Add a newline for better separation

# Another example with a different limit:
num = 12
display_multiplication_table(num, limit=5)
```

```python
def is_odd_or_even(number):
    """Checks if a given number is odd or even."""
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
num1 = 7
print(f"The number {num1} is {is_odd_or_even(num1)}.")

num2 = 10
print(f"The number {num2} is {is_odd_or_even(num2)}.")

num3 = 0
print(f"The number {num3} is {is_odd_or_even(num3)}.")
```

```python
def is_palindrome_number(number):
    """Checks if a given number is a palindrome."""
    # Convert the number to a string
    num_str = str(number)
    # Compare the string with its reverse
    return num_str == num_str[::-1]

# Example usage:
num1 = 121
print(f"The number {num1} is a palindrome: {is_palindrome_number(num1)}.")

num2 = 12345
print(f"The number {num2} is a palindrome: {is_palindrome_number(num2)}.")

num3 = 12321
print(f"The number {num3} is a palindrome: {is_palindrome_number(num3)}.")

num4 = -121 # Negative numbers are typically not considered palindromes in this context
print(f"The number {num4} is a palindrome: {is_palindrome_number(num4)}.")
```

```python
def collect_and_validate_student_data():
    """Collects and validates student input parameters."""

    # 1. Attendance Percentage (0-100)
    while True:
        try:
            attendance = int(input("Enter attendance percentage (0-100): "))
            if 0 <= attendance <= 100:
                break
            else:
                print("Invalid input. Attendance percentage must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter an integer for attendance percentage.")

    # 2. Internal Marks (0-30)
    while True:
        try:
            internal_marks = int(input("Enter internal marks (0-30): "))
            if 0 <= internal_marks <= 30:
                break
            else:
                print("Invalid input. Internal marks must be between 0 and 30.")
        except ValueError:
            print("Invalid input. Please enter an integer for internal marks.")

    # 3. Assignment Completion (yes/no)
    while True:
        assignment_completion = input("Has assignment completion been done (yes/no)? ").lower()
        if assignment_completion in ['yes', 'no']:
            break
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")

    # 4. Previous Semester CGPA (0.0-10.0)
    while True:
        try:
            cgpa = float(input("Enter previous semester CGPA (0.0-10.0): "))
            if 0.0 <= cgpa <= 10.0:
                break
            else:
                print("Invalid input. CGPA must be between 0.0 and 10.0.")
        except ValueError:
            print("Invalid input. Please enter a valid number for CGPA.")

    # 5. Backlogs (yes/no)
    while True:
        backlogs = input("Does the student have backlogs (yes/no)? ").lower()
        if backlogs in ['yes', 'no']:
            break
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")

    return attendance, internal_marks, assignment_completion, cgpa, backlogs

# Example usage of the function:
print("Please provide student data:")
validated_data = collect_and_validate_student_data()
print("\nValidated Student Data:")
print(f"  Attendance Percentage: {validated_data[0]}%")
print(f"  Internal Marks: {validated_data[1]}")
print(f"  Assignment Completion: {validated_data[2].capitalize()}")
print(f"  Previous Semester CGPA: {validated_data[3]}")
print(f"  Backlogs: {validated_data[4].capitalize()}")
```

```python
def initialize_output_structure():
    """Initializes the default structure for the expert system's outputs."""
    output_structure = {
        'Risk of failing': '',
        'Performance category': '',
        'Personalized suggestions': [],
        'Mentoring recommendation': ''
    }
    return output_structure

# Optional: Call the function and print the returned dictionary to verify its structure
print("Initialized Output Structure:")
initial_output = initialize_output_structure()
print(initial_output)
```

```python
def apply_expert_rules(attendance, internal_marks, assignment_completion, cgpa, backlogs):
    """Applies expert rules to student data to determine performance, risk, and recommendations."""

    output_structure = initialize_output_structure()

    # Convert boolean-like strings to actual booleans for easier rule application
    assignment_completed_bool = (assignment_completion == 'yes')
    has_backlogs_bool = (backlogs == 'yes')

    # --- Rule Set for Performance Category and Risk of Failing ---
    # Rule 1: High Risk & Weak Performance (based on critical factors)
    if has_backlogs_bool or attendance < 60 or internal_marks < 10 or cgpa < 5.0:
        output_structure['Risk of failing'] = 'HIGH'
        output_structure['Performance category'] = 'WEAK'
        if not has_backlogs_bool: # Add specific suggestion if backlogs not the primary reason for HIGH risk
            output_structure['Personalized suggestions'].append("Immediate intervention needed. Focus on core subjects and improving attendance.")

    # Rule 2: Medium Risk & Average Performance (balanced issues)
    elif (attendance < 75 and attendance >= 60) or \
         (internal_marks < 20 and internal_marks >= 10) or \
         (cgpa < 7.0 and cgpa >= 5.0) or \
         not assignment_completed_bool:
        output_structure['Risk of failing'] = 'MEDIUM'
        output_structure['Performance category'] = 'AVERAGE'
        output_structure['Personalized suggestions'].append("Review study habits and seek guidance from faculty or tutors.")

    # Rule 3: Low Risk & Good Performance
    else:
        output_structure['Risk of failing'] = 'LOW'
        output_structure['Performance category'] = 'GOOD'
        output_structure['Personalized suggestions'].append("Keep up the excellent work! Explore advanced topics or extracurricular activities.")

    # --- Personalized Suggestions (additional specific rules) ---
    # Rule 4: Specific attendance suggestion
    if attendance < 75:
        output_structure['Personalized suggestions'].append("Improve attendance to ensure you don't miss crucial lectures and discussions.")

    # Rule 5: Specific internal marks suggestion
    if internal_marks < 20:
        output_structure['Personalized suggestions'].append("Focus on preparing thoroughly for internal assessments and class tests.")

    # Rule 6: Specific assignment completion suggestion
    if not assignment_completed_bool:
        output_structure['Personalized suggestions'].append("Ensure timely submission of all assignments to gain continuous assessment points.")

    # Rule 7: Specific CGPA suggestion
    if cgpa < 7.0:
        output_structure['Personalized suggestions'].append("Work on understanding concepts better to improve your overall CGPA.")

    # Rule 8: Specific backlog suggestion
    if has_backlogs_bool:
        output_structure['Personalized suggestions'].append("Devote dedicated effort to clear all existing backlogs.")

    # --- Mentoring Recommendation ---
    # Rule 9: Mentoring for High/Medium risk or backlogs
    if output_structure['Risk of failing'] in ['HIGH', 'MEDIUM'] or has_backlogs_bool:
        output_structure['Mentoring recommendation'] = 'YES'
    # Rule 10: No mentoring otherwise
    else:
        output_structure['Mentoring recommendation'] = 'NO'

    # Remove duplicate suggestions by converting to a set and back to a list
    output_structure['Personalized suggestions'] = list(dict.fromkeys(output_structure['Personalized suggestions']))

    return output_structure

# Example usage with validated_data from previous step
print("\nApplying expert rules to sample data:")
sample_attendance, sample_internal_marks, sample_assignment_completion, sample_cgpa, sample_backlogs = validated_data
analysis_results = apply_expert_rules(sample_attendance, sample_internal_marks, sample_assignment_completion, sample_cgpa, sample_backlogs)

print("\nAnalysis Results for Sample Student:")
for key, value in analysis_results.items():
    if isinstance(value, list):
        print(f"  {key}:")
        for item in value:
            print(f"    - {item}")
    else:
        print(f"  {key}: {value}")

# Another example for a high-risk student
high_risk_results = apply_expert_rules(50, 8, 'no', 4.5, 'yes')
print("\nAnalysis Results for High-Risk Student:")
for key, value in high_risk_results.items():
    if isinstance(value, list):
        print(f"  {key}:")
        for item in value:
            print(f"    - {item}")
    else:
        print(f"  {key}: {value}")
```

```python
def display_results(analysis_output):
    """Displays the expert system's analysis and recommendations in a user-friendly format."""
    print("\n--- Student Performance Analysis ---")
    for key, value in analysis_output.items():
        if isinstance(value, list):
            print(f"{key}:")
            if value:
                for item in value:
                    print(f"  - {item}")
            else:
                print("  No specific suggestions at this time.")
        else:
            print(f"{key}: {value}")
    print("------------------------------------")

# Example usage with the 'analysis_results' from previous steps
print("\nDisplaying results for the first sample student:")
display_results(analysis_results)

# Example usage with the 'high_risk_results' from previous steps
print("\nDisplaying results for the high-risk student:")
display_results(high_risk_results)

# Example usage with a placeholder for a student with no suggestions
no_suggestions_data = {
    'Risk of failing': 'LOW',
    'Performance category': 'GOOD',
    'Personalized suggestions': [],
    'Mentoring recommendation': 'NO'
}
print("\nDisplaying results for a student with no specific suggestions:")
display_results(no_suggestions_data)
```
